```python
# ─────────────────────────────────────────────
# IMPORTS
# Ce qu'on importe et pourquoi :
#   pandas  → manipuler les données sous forme de tableau
#   numpy   → calculs mathématiques (moyenne, variance...)
#   plotly  → créer des graphiques interactifs
# ─────────────────────────────────────────────
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

print('✓ Librairies importées avec succès')
```

```python
# ─────────────────────────────────────────────
# CHARGER LE DATASET
# Adapte le chemin selon ton projet
# ─────────────────────────────────────────────
df = pd.read_csv('../data/unsw-nb15/UNSW_NB15_training-set.csv')

print('=' * 55)
print(f'  Lignes    : {df.shape[0]:,}')
print(f'  Colonnes  : {df.shape[1]}')
print(f'  Mémoire   : {df.memory_usage(deep=True).sum() / 1024**2:.1f} MB')
print('=' * 55)
```

```python
# ─────────────────────────────────────────────
# APERÇU DES PREMIÈRES LIGNES
# df.head() affiche les 5 premières lignes
# → permet de voir à quoi ressemble une connexion réseau
# ─────────────────────────────────────────────
print('5 premières lignes du dataset :')
df.head()
```

```python
# ─────────────────────────────────────────────
# TYPES DE CHAQUE COLONNE
# object  = texte (proto, service, state)
# int64   = nombre entier (spkts, dpkts...)
# float64 = nombre décimal (dur, rate...)
# ─────────────────────────────────────────────
print('Type de chaque colonne :')
print(df.dtypes.to_string())
```

```python
# ─────────────────────────────────────────────
# COMPTER LES CLASSES
# value_counts() compte combien de fois chaque valeur apparaît
# ─────────────────────────────────────────────
dist = df['label'].value_counts().reset_index()
dist.columns = ['label', 'count']
dist['nom']  = dist['label'].map({0: 'Normal (0)', 1: 'Attaque (1)'})
dist['pourcentage'] = (dist['count'] / dist['count'].sum() * 100).round(1)

print('Distribution des classes :')
print(dist.to_string(index=False))
print(f"\nRatio déséquilibre : {dist['count'].min() / dist['count'].max():.2%}")
print('→ Dataset relativement équilibré, pas besoin de sur-échantillonnage')
```

```python
# ─────────────────────────────────────────────
# GRAPHIQUE : Distribution des classes
# px.pie() → graphique camembert interactif
# ─────────────────────────────────────────────
fig = px.pie(
    dist,
    values='count',
    names='nom',
    title='Distribution des classes — Normal vs Attaque',
    color='nom',
    color_discrete_map={
        'Normal (0)' : '#1D9E75',
        'Attaque (1)': '#D85A30'
    },
    hole=0.4,
)
fig.update_traces(textposition='outside', textinfo='percent+label+value')
fig.show()
```

```python
# ─────────────────────────────────────────────
# DISTRIBUTION DES TYPES D'ATTAQUES
# ─────────────────────────────────────────────
attack_dist = df['attack_cat'].value_counts().reset_index()
attack_dist.columns = ['type', 'count']
attack_dist['pct'] = (attack_dist['count'] / len(df) * 100).round(2)

print('Types d\'attaques dans le dataset :')
print(attack_dist.to_string(index=False))
```

```python
# ─────────────────────────────────────────────
# GRAPHIQUE : Types d'attaques
# px.bar() → graphique en barres horizontales
# log_x=True → échelle logarithmique car Worms (44) vs Generic (18871)
# ─────────────────────────────────────────────
fig = px.bar(
    attack_dist.sort_values('count'),
    x='count', y='type',
    orientation='h',
    title='Distribution des types d\'attaques (échelle log)',
    text='count',
    color='count',
    color_continuous_scale='Blues',
    log_x=True,
)
fig.update_traces(textposition='outside')
fig.update_layout(height=450, showlegend=False)
fig.show()

print('\nObservation importante :')
print('→ Worms (44 exemples) est très rare : le modèle aura du mal à le détecter')
print('→ Generic (18871) est très fréquent : le modèle sera bon dessus')
print('→ Pour notre projet on utilise label (0/1), pas attack_cat')
```

```python
# ─────────────────────────────────────────────
# DÉTECTER LES VALEURS MANQUANTES
# isnull()  → True si la valeur est vide
# sum()     → compte les True (= les vides)
# ─────────────────────────────────────────────
manquants = df.isnull().sum()
nb_total  = df.isnull().sum().sum()

print(f'Total valeurs manquantes : {nb_total}')

if nb_total == 0:
    print('✓ Aucune valeur manquante — dataset propre !')
    print('→ Pas besoin d\'imputation, on peut passer directement au prétraitement')
else:
    cols_manquantes = manquants[manquants > 0].sort_values(ascending=False)
    print(f'\n{len(cols_manquantes)} colonnes avec des valeurs manquantes :')
    print(cols_manquantes)
```

```python
# ─────────────────────────────────────────────
# CORRÉLATION AVEC LE LABEL
# .corr()  → matrice de corrélation entre toutes les colonnes
# ['label'] → on garde seulement la colonne label
# .abs()   → valeur absolue (corrélation négative = aussi utile)
# ─────────────────────────────────────────────
cols_num = df.select_dtypes(include=[np.number]).columns.tolist()
cols_num = [c for c in cols_num if c not in ['id', 'label']]

corr = df[cols_num + ['label']].corr()['label'].drop('label').abs().sort_values(ascending=False)
top15 = corr.head(15).reset_index()
top15.columns = ['feature', 'correlation']

print('Top 15 features les plus corrélées avec le label :')
for _, row in top15.iterrows():
    barre = '█' * int(row['correlation'] * 30)
    print(f"  {row['feature']:<25} {barre} {row['correlation']:.4f}")
```

```python
# ─────────────────────────────────────────────
# GRAPHIQUE : Top 15 features
# ─────────────────────────────────────────────
fig = px.bar(
    top15.sort_values('correlation'),
    x='correlation', y='feature',
    orientation='h',
    title='Top 15 features les plus prédictives (corrélation avec label)',
    text=top15.sort_values('correlation')['correlation'].round(3),
    color='correlation',
    color_continuous_scale='Blues',
)
fig.update_traces(textposition='outside')
fig.update_layout(height=500, xaxis_range=[0, 0.6])
fig.show()

print('\nObservations :')
print(f'→ Feature la plus prédictive : {top15.iloc[0]["feature"]} (corr={top15.iloc[0]["correlation"]:.3f})')
print(f'→ 2ème : {top15.iloc[1]["feature"]} (corr={top15.iloc[1]["correlation"]:.3f})')
print('→ Ces features seront probablement les plus utilisées par le Hoeffding Tree')
```

```python
# ─────────────────────────────────────────────
# COLONNES CATÉGORIELLES
# ─────────────────────────────────────────────
cols_cat = ['proto', 'service', 'state']

fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=['Protocole (proto)', 'Service', 'État (state)']
)

couleurs = ['#378ADD', '#1D9E75', '#7F77DD']
for i, col in enumerate(cols_cat):
    top = df[col].value_counts().head(8)
    fig.add_trace(
        go.Bar(x=top.index, y=top.values, marker_color=couleurs[i],
               name=col, showlegend=False),
        row=1, col=i+1
    )

fig.update_layout(title='Distribution des features catégorielles', height=380)
fig.show()

for col in cols_cat:
    n_unique = df[col].nunique()
    print(f'{col}: {n_unique} valeurs uniques → à encoder avec pd.Categorical().codes')
```

```python
# ─────────────────────────────────────────────
# HISTOGRAMMES DES 9 FEATURES LES PLUS CORRÉLÉES
# On compare Normal vs Attaque pour chaque feature
# ─────────────────────────────────────────────
top9_features = top15['feature'].head(9).tolist()

fig = make_subplots(rows=3, cols=3, subplot_titles=top9_features)
df_normal  = df[df['label'] == 0]
df_attaque = df[df['label'] == 1]

for i, feat in enumerate(top9_features):
    row, col_idx = i // 3 + 1, i % 3 + 1
    fig.add_trace(
        go.Histogram(x=df_normal[feat], name='Normal',
                     marker_color='#1D9E75', opacity=0.6,
                     showlegend=(i == 0), nbinsx=40),
        row=row, col=col_idx
    )
    fig.add_trace(
        go.Histogram(x=df_attaque[feat], name='Attaque',
                     marker_color='#D85A30', opacity=0.6,
                     showlegend=(i == 0), nbinsx=40),
        row=row, col=col_idx
    )

fig.update_layout(
    title='Distribution des top 9 features : Normal vs Attaque',
    barmode='overlay',
    height=600,
    legend=dict(orientation='h', y=1.05)
)
fig.show()

print('Interprétation :')
print('→ Si les 2 distributions sont séparées : la feature est très prédictive')
print('→ Si elles se superposent : la feature est moins utile')
```
